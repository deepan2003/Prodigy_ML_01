```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"D:\Datasets\house-prices-advanced-regression-techniques\train.csv")
data
```

```python
data.isnull().sum()
```

Removing null values

```python

col = "LotFrontage"
for i in col:
    data[col] = data[col].fillna(data[col].mean())
data[col]
```

```python
data.isnull().sum()
```

Replacing Nan values as 0

```python
col = 'Alley','PoolQC','Fence','MiscFeature'
for i in col:
    data[i] = data[i].replace(np.nan,0)
data[i]
```

```python
data
```

```python
x = data[["1stFlrSF","2ndFlrSF","LowQualFinSF"]]
y = data[["BedroomAbvGr"]]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
lr = LinearRegression()
```

```python
fitting = lr.fit(x_train,y_train)
fitting
```

```python
prediction = lr.predict(x_test)
prediction
```

```python
slope = lr.coef_
slope
```

```python
intercept = lr.intercept_
intercept
```

```python
mse=mean_squared_error (y_test , prediction)
mse
```
